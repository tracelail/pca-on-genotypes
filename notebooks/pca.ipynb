```python
import pandas as pd
from sklearn import decomposition
import matplotlib.pyplot as plt
import altair as alt
import os
```

```python
current_dir = os.getcwd()
input_path = os.path.abspath(os.path.join(current_dir, "../data"))
output_path = os.path.abspath(os.path.join(current_dir, "../results"))
```

```python
matrix = pd.read_csv(os.path.join(output_path, "matrix.csv"), index_col=0)
matrix
```

```python
pca = decomposition.PCA(n_components=2)
pca.fit(matrix)
print(pca.explained_variance_ratio_)
print(pca.singular_values_)
pca_matrix = pca.transform(matrix)
pca_matrix.shape
```

```python
pca_matrix
```

```python
pc1 = pca_matrix[:, 0]
pc2 = pca_matrix[:, 1]
```

```python
sample_metadata = pd.read_csv(os.path.join(output_path, "sample-metadata.csv"))
sample_metadata
```

```python
to_plot = sample_metadata.copy()
to_plot["PC1"] = pc1
to_plot["PC2"] = pc2
to_plot
```

# PCA plotting

```python
alt.Chart(to_plot).mark_point().encode(
    x="PC1", y="PC2", color="Superpopulation name"
).interactive()
```
