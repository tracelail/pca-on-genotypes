```python
import pandas as pd
```

```python
matrix = pd.read_csv("../results/matrix.csv", index_col=0)
matrix
```

```python
panel_filename = "../data/phase1_integrated_calls.20101123.ALL.panel"

with open(panel_filename) as panel_file:
    labels = {}  # {sample_id : population_code}
    for line in panel_file:
        line = line.strip().split("\t")
        labels[line[0]] = line[1]

print(list(labels.items())[:5])
```

```python
matrix_pop_codes = matrix.copy()
matrix_pop_codes["population_code"] = matrix_pop_codes.index.map(labels)
matrix_pop_codes = matrix_pop_codes.reset_index().rename(columns={"index": "sample"})
matrix_pop_codes
```

```python
igsr_pop = pd.read_csv("../data/igsr_populations.tsv", sep="\t")
igsr_pop
```

```python
matrix_metadata = matrix_pop_codes.merge(
    igsr_pop, left_on="population_code", right_on="Population code"
)
matrix_metadata
```

```python
matrix_metadata.to_csv("../results/matrix-metadata.csv", index=False)
```

```python
sample_pop_codes = matrix_pop_codes[["sample", "population_code"]]
sample_metadata = sample_pop_codes.merge(
    igsr_pop, left_on="population_code", right_on="Population code"
).drop(columns=["population_code"])
sample_metadata
```

```python
sample_metadata.to_csv("../results/sample-metadata.csv", index=False)
```
